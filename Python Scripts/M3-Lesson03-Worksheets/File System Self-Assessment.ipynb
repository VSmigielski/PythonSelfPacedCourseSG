```python
# Veronica Smigielski 26/04/2021
class FileItem:
    permissions = ""
    owner = ""
    size = ""
    def __init__(self, p,o,s):
        self.permissions = p
        self.owner = o
        self.size = s
    def display(self):
        print("Permissions: " + self.permissions)
        print("Owner: " + self.owner)
        print("Size:", self.size)
class CsvFile(FileItem):
    contains = ""
    application = ""
    def __init__(self, p,o,s,c,a):
        FileItem.__init__(self, p, o, s)
        self.contains = c
        self.application = a
    def display(self):
        print("Permissions: " + self.permissions)
        print("Owner: " + self.owner)
        print("Size:", self.size)
        print("Type: " + self.contains)
        print("Application to Open: " + self.application)
class JpgFile(FileItem):
    picture_of = ""
    application = ""
    def __init__(self, p,o,s,pa,a):
        FileItem.__init__(self, p, o, s)
        self.picture_of = pa
        self.application = a
    def display(self):
        print("Permissions: " + self.permissions)
        print("Owner: " + self.owner)
        print("Size:", self.size)
        print("Picture Of: " + self.picture_of)
        print("Application to Open: " + self.application)
class Mp3File(FileItem):
    music_type = ""
    application = ""
    def __init__(self, p,o,s,mt,a):
        FileItem.__init__(self, p, o, s)
        self.music_type = mt
        self.application = a
    def display(self):
        print("Permissions: " + self.permissions)
        print("Owner: " + self.owner)
        print("Size:", self.size)
        print("Music Type: " + self.music_type)
        print("Application to Open: " + self.application)
class Directory(FileItem):
    parent = None
    child = None
    def __init__(self, p, o, s, pa, *c):
        FileItem.__init__(self, p, o, s)
        self.parent = pa
        self.child = c
    def display(self):
        print("Directory: " + self.permissions + "/" + self.owner + "/" + self.parent)
        for i in self.child:
            if isinstance(i, Mp3File):
                Mp3File.display(i)
            elif isinstance(i, JpgFile):
                JpgFile.display(i)
            elif isinstance(i, CsvFile):
                CsvFile.display(i)
            
c1 = CsvFile("Admin", "Vero", 10, "Bath Bombs", "Microsoft Excel")
c2 = CsvFile("User", "Gio", 20, "Meeting Info", "Google Drive")
c3 = CsvFile("User", "Gio", 30, "Work", "Microsoft Excel")
c4 = CsvFile("Admin", "Vero", 48, "Soap", "Microsoft Excel")
c5 = CsvFile("Admin", "Vero", 36, "Pokemon Project", "Google Drive")

c1.display()
c2.display()
c3.display()
c4.display()
c5.display()

j1 = JpgFile("Admin", "Vero", 2, "Paris", "Photos")
j2 = JpgFile("User", "Gio", 4, "Amsterdam", "Instagram")
j3 = JpgFile("User", "Gio", 5, "Brugge", "Photos")
j4 = JpgFile("Admin", "Vero", 8, "Snow in Belgium", "Photos")
j5 = JpgFile("Admin", "Vero", 7, "Amsterdam", "Facebook")

j1.display()
j2.display()
j3.display()
j4.display()
j5.display()
            
m1 = Mp3File("Admin", "Vero", 100, "Rock", "Windows Play")
m2 = Mp3File("User", "Gio", 120, "Classic", "Windows Play")
m3 = Mp3File("User", "Gio", 180, "EDM", "Website")
m4 = Mp3File("Admin", "Vero", 240, "EDM", "Apple Play")
m5 = Mp3File("Admin", "Vero", 200, "Party", "Windows Play")

m1.display()
m2.display()
m3.display()
m4.display()
m5.display()
        
dir1 = Directory("Music", "Public", 0, "Vero", m1, m3, m4)
dir2 = Directory("Photos", "Public", 0, "Vero", j2, j5)
dir3 = Directory("Documents", "Public", 0, "Gio", c3, c4)
dir4 = Directory("Music", "Public", 0, "Vero", m2, m4, m5)
dir5 = Directory("Photos", "Public", 0, "Gio", j1, j3, j4)

dir1.display()
dir2.display()
dir3.display()
dir4.display()
dir5.display()
```
