Implement the classes below as follows:
* Implement the `__init__` and `display` methods of the `Asset` class.

    * Implement the `__init__` and `display` methods of the `Stock` class.

* Create two objects from each of the `Stock`, `Bond`, and `MutualFund` classes. Use the `display` method to display the contents of each object.  

```python
class Asset: 
    symbol=None
    price=None
    description=None
    def __init__(self,sym,pce,desc):
        self.symbol = sym
        self.price = pce
        self.description = desc
    def display(self):
        print("Asset Symbol: " + self.symbol)
        print("Asset Price: " + self.price)
        print("Asset Description: " + self.description)
    
class Stock(Asset):
    company_name = None
    def __init__(self, sym, pce, desc, comp):
        Asset.__init__(self, sym, pce, desc)
        self.company_name = comp
    def display(self):
        Asset.display(self)
        print("Company name: " + self.company_name)

class Bond(Asset):
    interest_rate=None
    def __init__(self, sym, pce, desc, rate):
        Asset.__init__(self, sym, pce, desc)
        self.interest_rate = rate
    def display(self):
        Asset.display(self)
        print("Interest rate: " + self.interest_rate)

class MutualFund(Asset):
    fund_manager=None
    def __init__(self, sym, pce, desc, man):
        Asset.__init__(self, sym, pce, desc)
        self.fund_manager = man
    def display(self):
        Asset.display(self)
        print("Fund manager: " + self.fund_manager)

s1 = Stock("Star", "42.99", "This is a star of a fun time", "StarsUnited")
s2 = Stock("Heart", "109.99", "This is the heart of freedom", "Hearts Aligned")

b1 = Bond("Star", "42.99", "This is a star of a fun time", "1.3")
b2 = Bond("Heart", "109.99", "This is the heart of freedom", "4.2")

m1 = MutualFund("Star", "42.99", "This is a star of a fun time", "Gio")
m2 = MutualFund("Heart", "109.99", "This is the heart of freedom", "Veronica")

s1.display()
s2.display()

b1.display()
b2.display()

m1.display()
m2.display()
```
