```python
import string

# Declare a string to use user input
s = input("Please enter a string to be converted to a list, set, and dictionary:")

# lowercase the string
s_lower = s.lower()

# Split the string into a list of singular words
words = list(s_lower.split())

# Make a new set to hold the words (no duplicates)
words_set = set()

# Initialize count for list
count_list = 0

# Use a for loop to get the list count
for i in words:
    count_list += 1
    
# Use a for loop to insert the list words into the set
for word in words:
    words_set.add(word)
    
# Count the words in the set
count_set = len(words_set)

# Create a new dictionary to be able to count the instances of each word inside the list
word_freq = dict()

# This list is the one with punctuation
for word in words:
    word_freq[word] = (words.count(word))
    
# Use the string import to get a list of punctuation
punctuation_list =  list(string.punctuation)

# New list to hold those without punctuation
w_clean = list()

# For loop to check for length greater than 1 & to eliminate the punctuation
# in the first and last index of each word
for word in words:
    new_word = ""
    for char in word:
        if char[0] and char[-1] not in punctuation_list:
            new_word += char
            
    if len(new_word) > 0:
        w_clean.append(new_word)

# Count the words in the clean list
count_clean = len(w_clean)

# Prompt the user for a word to check its frequency
user_freq = input("Which word would you like to see the frequency for? ")
output = dict()

# Check the key values from user's input, insert the key and value into a new
# dictionary
for k, v in word_freq.items():
    if k == user_freq:
        output[k] = v

# Sort dictionary of frequency alphabetically
sorted_dict_alpha = dict(sorted(word_freq.items(), key=lambda item: item[0]))

# Print all of the variables with proper explanation for the user
print("The initial string:\n", s, "\n")
print("The lower cased string:\n", s_lower, "\n")
print("The list containing the punctuation from the string:\n", words, "\n")
print("The word count for the list is:\n", count_list, "\n")
print("The set containing the punctuation after list is inserted:\n", words_set, "\n")
print("The word count for the set is:\n", count_set, "\n")
print("The dictionary containing the distinct words and count frequency per word:\n", word_freq, "\n")
print("This is the list that does not contain punctuation in the first and last index:\n", w_clean, "\n" )
print("This is the count for the clean list:\n", count_clean, "\n")
print("This is the key:value pair that you searched for:\n", output, "\n")
print("This is the frequency dictionary in alphabetical order:\n", sorted_dict_alpha, "\n")

# Sort dictionary of frequency by frequency lowest to highest
print("This is the dictionary sorted from lowest to highest frequency:\n", {key: value for key, value in sorted(word_freq.items(), key=lambda item: item[1])})
```

```python
def fib2(num):
    a, b = 1, 1
    for _ in range(num - 1):
        a, b = b, a + b
    return a  
print(fib2(7))
```
