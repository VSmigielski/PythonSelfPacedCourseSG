## Activity 1

Create a script that uses the `map` function to compute the square root of each value in an input set of integers.

```python
# your code here
import math

list_numbers = []

user_input = ""

while user_input.lower() != "stop":
    user_input = input("Enter a number or 'stop' to end the program:\n")
    if user_input.lower() != "stop":
        list_numbers.append(int(user_input))
    else:
        break
        
map_iterator = map(lambda x : math.sqrt(x) , list_numbers)

map_list = list(map_iterator)
print(map_list)
```

## Activity 3

Create a script that uses the `map` function to determine which integers are prime numbers in an input list of integers.

```python
# your code here

def find_prime(val):
    x = val
    for num in range(2, x - 1):
        if val % num == 0:
            x = 0
    if x != 0:
        return True
    else:
        return False

list_numbers = []

user_input = ""

while user_input.lower() != "stop":
    user_input = input("Enter a number or 'stop' to end the program:\n")
    if user_input.lower() != "stop":
        list_numbers.append(int(user_input))
    else:
        break
        
numbers_prime = map(find_prime, list_numbers)

num_prime_list = list(numbers_prime)

numbers_dict = dict(zip(list_numbers, num_prime_list))

print(numbers_dict)
```

## Activity 8

Create a script that uses the `reduce` function to identify the key in a dictionary that is last in alphabetical order.

Example input:

    dict_my = {"elephant":1, "zebra":4, "panther":5}

Output:

    zebra

In this case, "zebra" is the key that appears last in alphabetical order.

```python
# your code here
from functools import reduce
dict_my = {"elephant":1, "zebra":4, "panther":5}

last_alphabet = reduce(lambda a,b : a if a[0] > b[0] else b,dict_my)

print(last_alphabet)
```


## Activity 10

Create a script that uses the `reduce` function to compute the sum of all even numbers in a list of numbers.

For example, given the list `[3, 4, 5, 9, 6]`, the output should be `10`. (4 and 6 are the only even numbers in the list, and 4 + 6 = 10.)

```python
# your code here
from functools import reduce

num_list = [3, 4, 5, 9, 6]

even_list = []

for a in num_list:
    if a % 2 == 0:
        even_list.append(a)
    else:
        pass

evens = reduce(lambda a, b : a + b, even_list)

print(evens)
```

## Activity 11

Create a script that uses the `reduce` function to compute the sum of all positive numbers within a list of numbers.

For example, the input `[1, 2, -1, -2, 1]` should output `4` (the sum of 1 + 2 + 1).

```python
# your code here
from functools import reduce

num_list = [1, 2, -1, -2, 1]

pos_list = []

for a in num_list:
    if a > 0:
        pos_list.append(a)
    else:
        pass

positive = reduce(lambda a, b : a + b, pos_list)

print(positive)
```

## Activity 12

Create a script that uses the `reduce` function to compute the sum of all negative numbers within a list of numbers.

For example, `[1,2,-1,-2,1]` should output `-3` (the sum of -1 + -2).

```python
# your code here
from functools import reduce

num_list = [1, 2, -1, -2, 1]

neg_list = []

for a in num_list:
    if a < 0:
        neg_list.append(a)
    else:
        pass

negative = reduce(lambda a, b : a + b, neg_list)

print(negative)
```
