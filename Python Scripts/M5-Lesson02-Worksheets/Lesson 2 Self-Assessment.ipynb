```python
# Veronica Smigielski 08/05/2021
import math
from functools import reduce
def fromCSV(path,delimiter,quotechar):
    import csv # import the csv module
    data=list() # any data we will read will be returned in a list 
    with open(path, newline='') as csvfile: # open the file
        filecontent = csv.DictReader(csvfile, delimiter=delimiter, quotechar=quotechar) 
        # access the content of the file
        for row in filecontent:# iterate through the rows 
            data.append(row) # save the rows in the data list. Each row is a dictionary 
    return data

def find_make(dataset):
    spec_make = input("Enter a make to look for: ")
    cars_make_filtered = filter(lambda x: x['make'] == spec_make, dataset) 
    cars_make_filtered_list = list(cars_make_filtered)

    return cars_make_filtered_list

def find_fuel_efficient(dataset):
    cars_fe_filtered = filter(lambda x : True if int(x['city_mpg']) > 35 else False, dataset) 
    cars_fe_filtered_list = list(cars_fe_filtered)

    return cars_fe_filtered_list

def find_horsepower(dataset):
    cars_hp_filtered = filter(lambda x : True if x['horsepower'] != "?" and float(x['horsepower']) > 100 else False, dataset) 
    cars_hp_filtered_list = list(cars_hp_filtered)

    return cars_hp_filtered_list

def display_cost_miles(cost, row):
    # input is the entire row from data
    # extract the month from the date field 
    # add the month field to the row and return the row 
    value = row['city_mpg']
    mpg = ""
    mpg_str = value
    new_row = row.copy()
    cost_rounded = round(((100 / float(mpg_str)) * cost), 2)
    new_row.update({'Cost per 100 miles': cost_rounded })
    return new_row

def reduce_avg(name, dataset):
    count = 0
    
    sum_list = []
    
    for row in dataset:
    #remove pass and add code here   
        if row == name:
            pass
        else:
            if row[name] == "?":
                count += 1
            else:
                sum_list.append(float(row[name]))
                count += 1
            
    sum_of_list = reduce(lambda x, y: x + y,sum_list)
    
    avg = round((sum_of_list / count), 2)
    
    return avg

#1
data=fromCSV(path='car_data/car_data.csv',delimiter=',',quotechar='|') 

#2
c_spec_make = find_make(data)
print(c_spec_make)
print("\n")

#3
c_fuel_efficient = find_fuel_efficient(data)
print(c_fuel_efficient)
print("\n")

#4
c_horsepower = find_horsepower(data)
print(c_horsepower)
print("\n")

#5
cost = float(input("Enter the cost of gas per gallon:\n"))
data_mapped = map(lambda x : display_cost_miles(cost, x),data)
data_mapped_list = list(data_mapped)
print(data_mapped_list)
print("\n")

#6
city_mpg_avg = reduce_avg("city_mpg", data)
print("The average city mpg for all cars is:", city_mpg_avg)
print("\n")
hwy_mpg_avg = reduce_avg("highway_mpg", data)
print("The average highway mpg for all cars is:", hwy_mpg_avg)
print("\n")

#7
price_avg = reduce_avg("price", data)
print("The average price for all cars is:", price_avg)
```
