## Activity 1
Create an iterator class named `PrimeNumbers` that creates a sequence of prime numbers between 0 and 100. 

```python
# your code here
class PrimeNumbers:
    # remove pass and add code here
    def __iter__(self):
        self.a = 0
        return self
    def __next__(self):
        if (self.a < 101):
            flag = False
            x = self.a
            while (flag == False and self.a < 101):
                x = self.find_prime(self.a)
                self.a += 1
                if x != 0:
                    flag = True
                    
            if self.a > 100:
                pass
            else:
                return x
            
        raise StopIteration
        
    def find_prime(self, val):
        x = val
        for num in range(2, x - 1):
            if val % num == 0:
                x = 0
                break
        return x
        
numbers = PrimeNumbers()
numbers_iterator = iter(numbers)
for number in numbers_iterator: 
    print(number)
```

## Activity 2
Create an iterator class named `Alphabet` that creates a sequence of characters representing the alphabet.

```python
# your code here
from string import ascii_lowercase
class Alphabet:
    # remove pass and add your code here
    def __iter__(self):
        self.a = 65
        return self
    def __next__(self):
        if self.a < 91:
            x = chr(self.a).lower()
            self.a += 1
            return x
        raise StopIteration

alphabet = Alphabet()
alphabet_iterator = iter(alphabet)
for alpha in alphabet_iterator: 
    print(alpha)
```

## Activity 3
Create an iterator class named `DivBy5` that creates a sequence of numbers between 0 and 1000 inclusive, where each number is a multiple of 5. 

```python
# your code here
class DivBy5:
    # remove pass and add your code here
    def __iter__(self):
        self.a = 0
        return self
    def __next__(self):
        if (self.a < 1001):
            x = self.a
            self.a += 5
            return x
        raise StopIteration

numbers = DivBy5()
numbers_iterator = iter(numbers)
for number in numbers_iterator: 
    print(number)
```

## Activity 4
Create an iterator class named `Factorial` that creates a sequence of numbers representing the factorial of all numbers between 0 and 100. 

```python
# your code here
class Factorial:
    # remove pass and add your code here
    def __iter__(self):
        self.a = 0
        return self
    def __next__(self):
        if (self.a < 101):
            x = self.a
            if (x == 0 or x == 1):
                self.a += 1
                return x
            else:
                fac = self.a
                while (fac != 1):
                    x *= (fac - 1)
                    fac -= 1
                self.a += 1
                return x
        raise StopIteration

numbers = Factorial()
numbers_iterator = iter(numbers)
for number in numbers_iterator: 
    print(number)
```

## Activity 5
Create an iterator class that generates a sequence of random characters from the alphabet.  

```python
# your code here
import random
class Alphabet:
    # remove pass and add your code here
    def __iter__(self):
        self.a = 0
        return self
    def __next__(self):
        if self.a < 10:
            x = chr(random.randint(1, 1000))
            self.a += 1
            return x
            
        raise StopIteration

alphabet = Alphabet()
alphabet_iterator = iter(alphabet)
for alpha in alphabet_iterator: 
    print(alpha)
```
