```python
# Veronica Smigielski 27/04/2021
class FoodItem:
    name = ""
    price = ""
    def __init__(self, n, p):
        self.name = n
        self.price = p
    def display(self):
        print("Name of Item: " + self.name)
        print("Price of Item: ", self.price)
        
class Burger(FoodItem):
    condiments = ""
    def __init__(self, n, p, c):
        FoodItem.__init__(self, n, p)
        self.condiments = c
    def display(self):
        print("Name of Item: " + self.name)
        print("Price of Item: ", self.price)
        print("Condiments: " + self.condiments)
        
class Drink(FoodItem):
    kind = ""
    size_drink = ""
    def __init__(self, n, p, k, s):
        FoodItem.__init__(self, n, p)
        self.kind = k
        self.size_drink = s
    def display(self):
        print("Name of Item: " + self.name)
        print("Price of Item: ", self.price)
        print("Kind: " + self.kind)
        print("Size: " + self.size_drink)
        
class Side(FoodItem):
    size = ""
    def __init__(self, n, p, s):
        FoodItem.__init__(self, n, p)
        self.size = s
    def display(self):
        print("Name of Item: " + self.name)
        print("Price of Item:", self.price)
        print("Size: " + self.size)
        
class Combo(FoodItem):
    burger = ""
    drink = ""
    side = ""
    def __init__(self, n, p, b, d, s):
        FoodItem.__init__(self, n, p)
        self.burger = b
        self.drink = d
        self.side = s
    def display(self):
        print("Burger Display: " + Burger.display(self))
        print("Drink Display: " + Drink.display(self))
        print("Side Display: " + FoodItem.display(self))
        
class Order:
    name = ""
    list_of_food = []
    def __init__(self, n, l):
        self.name = n
        self.list_of_food = l
    def display(self):
        print("Name of Customer: " + self.name)
        for i in self.list_of_food:
            i.display()
#            if isinstance(i, Burger):
#                Food.display(i)
#            elif isinstance(i, Drink):
#                Drink.display(i)
#            elif isinstance(i, Side):
#                Side.display(i)
#            elif isinstance(i, Combo):
#                Combo.display(i)
 
def user_input_burger():
    #ask user for input and store it in burger object 
    user_input = input("What condiments would you like with the burger?")
    b = Burger("Hamburger", 6.50, user_input)
    return b
 
def user_input_drink():
    #ask user for input and store it in drink object 
    user_input_kind = input("What kind of drink would you like?")
    user_input_size = input("What size drink would you like? \nsmall, medium, large, or x-large: ")
    if user_input_size.lower() == "small":
        d = Drink("Drink", 2.00, user_input_kind, user_input_size)
    elif user_input_size.lower() == "medium":
        d = Drink("Drink", 2.50, user_input_kind, user_input_size)
    elif user_input_size.lower() == "large":
        d = Drink("Drink", 3.00, user_input_kind, user_input_size)
    elif user_input_size.lower() == "x-large":
        d = Drink("Drink", 3.50, user_input_kind, user_input_size)
    return d
 
def user_input_side():
    user_input_size = input("What size side would you like? \nsmall, medium, large, or x-large: ")
    #ask user for input and store it in side object 
    if user_input_size.lower() == "small":
        s = Side("Fries", 1.50, user_input_size)
    elif user_input_size.lower() == "medium":
        s = Side("Fries", 2.50, user_input_size)
    elif user_input_size.lower() == "large":
        s = Side("Fries", 3.50, user_input_size)
    elif user_input_size.lower() == "x-large":
        s = Side("Fries", 4.50, user_input_size)
    return s
 
def user_input_combo():
    b = user_input_burger()
    d = user_input_drink()
    s = user_input_side()
        
    c = Combo(b, d, s)
    #ask user for input and store it in combo object 
    #a combo must include one burger, one side, and one drink
    return c
 
def take_order():
    #ask user for name for the order 
    #repeat taking order until client is done 
    #display order details
    #display a thank you message
    print("Welcome to Burger Shop")
    list_of_food = []
    
    name_input = input("What is your name? ")
    user_input = ""
    
    while user_input.lower() != "quit" or user_input.lower() != "cancel":
        user_input = input("What would you like to order? \nThe menu options are burger, drink, side, or combo")
        if user_input.lower() == "quit" or user_input.lower() == "cancel":
            break
        elif user_input.lower() == "burger":
            new = user_input_burger()
            list_of_food.append(new)
        elif user_input.lower() == "drink":
            new = user_input_drink()
            list_of_food.append(new)
        elif user_input.lower() == "side":
            new = user_input_side()
            list_of_food.append(new)
        elif user_input.lower() == "combo":
            new = user_input_combo()
            list_of_food.append(new)
        else: 
            print("That was not an option. Please try again.")
            
    o = Order(name_input, list_of_food)
    o.display()
    print("Thank you for your order.")
    
take_order()
```
