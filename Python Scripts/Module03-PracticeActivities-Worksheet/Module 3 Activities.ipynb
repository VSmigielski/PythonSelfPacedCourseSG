## Activity 1
Write a program that performs the following tasks:
1. Accept at least five and no more than ten integers from the user and store the values in a list.
    * Give the user the option to stop input after receiving five numbers.
    * Stop accepting input when the user chooses to stop or when there are ten numbers in the list.
1. Calculate the average value of the input numbers and use this value as a threshold.
1. Create a `find_lower` function that accepts the list of numbers and the threshold value as parameters and then calculates and displays the values from the list that are lower than the threshold value.

```python
# your code here

def find_lower(numbers,threshold):
    for num in numbers:
        # we only display the numbers that are above the input threshold
        if (num < threshold):
            print(num)
def compute_sum(numbers):
    # remove pass and add your code here
    sum = 0
    for i in numbers:
        sum += i
    return sum

numbers = []
userInput = ""
count = 0
while count < 10:
    userInput = input("Enter an integer: ")
    if count < 5:
        user_int = int(userInput)
        numbers.append(user_int)
    elif count >= 5:
        if userInput.lower() == "quit":
            break
        else:
            user_int = int(userInput)
            numbers.append(user_int)
    count += 1
    
summation = compute_sum(numbers)
len_of_num = len(numbers)
average = summation/len_of_num
print("This is the average:", average)
print("These are the numbers lower than the average from the list:")
find_lower(numbers,average) # find all numbers lower than the average
```

## Activity 2
Create a function that searches a string for an input group of characters and replaces each instance of the group with a new group of characters.

For example, the function might be used to replace all occurrences of the phrase "this is" with "this will be."

```python
# your code here

def replace_words(text):
    text_r = text.replace("this is", "this will be")
    print(text_r)
    
text = "this is a long day. this is a short function. this is a test."
replace_words(text)

def replace_words(ow, rw, text):
    import re
    old_words = [ow]
    big_regex = re.compile('|'.join(map(re.escape, old_words)))
    the_message = big_regex.sub(rw, text)
    print(the_message)

ow = 'this is'
rw = 'this will be'
s = 'this is a long day. this is a short function. this is a test.'
replace_words(ow, rw, s)
```
