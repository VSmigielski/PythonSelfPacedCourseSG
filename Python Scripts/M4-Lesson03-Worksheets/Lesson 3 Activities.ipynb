## Activity 5
Write a function that takes as input the path to a text file and returns the longest word in the text file.

```python
# your code here
def longest_word(filename):
    # remove pass and add code here
    with open(filename, 'r') as f:
        words = f.read().split()
    max_len = len(max(words, key=len))
    
    return [word for word in words if len(word) == max_len]

user_input = input("Enter a path and filename. Ex: FileIO-DataFiles/flatland01.txt\n")

word = longest_word(user_input)

print(word)
```

## Activity 6
Write a function that takes as input a list of strings and writes the list to a text file. 

Each element in the list should be a separate line in the new text file.

```python
# your code here
def list_to_file(items, user_input):
    with open(user_input, 'w') as f:
        for listitem in items:
            f.write('%s\n' % listitem)
        
    f.close()
    return user_input

item_list = []

user_input_items = ""

user_input = input("Enter a path and filename where you want to save the information. Ex: FileIO-DataFiles/flatland01.txt\n")

while user_input_items.lower() != "quit":
    user_input_items = input("Enter a few items. \n")
    if user_input_items.lower() != "quit":
        item_list.append(user_input_items)
    else:
        break
        
new_file = list_to_file(item_list, user_input)

f = open(new_file, "r")  

print("\n")
print(f.read())

f.close() 
```

## Activity 7
Write a function that takes as input a text file and a new filename, and then copies the contents of the input text file to another file using the new filename.

**Challenge**: Update the code so that the lines in the new file are in the reverse order of the lines in the original file.

```python
# your code here
def file_to_file(filepath1, filepath2):
    with open(filepath2, 'w') as outfile:
        with open(filepath1) as infile:
            lines = infile.readlines()
            for line in reversed(lines):
                outfile.write(line)
            
    outfile.close()
    
    return filepath2

filepath1 = input("Enter a path and filename where you want to retrieve information. Ex: FileIO-DataFiles/flatland01.txt\n")
filepath2 = input("Enter a path and filename where you want to save the information. Ex: FileIO-DataFiles/flatland01.txt\n")

file_created = file_to_file(filepath1, filepath2)

f = open(file_created, "r")  

print(f.read())

f.close() 
```

The code should be as automated as possible, so that the user simply has to identify the original file and does not need to know how many lines of text the file includes.
